# Chapter 18 — Experiment Tracking with W&B

Training a model is just the beginning — managing it in production is where the real engineering challenge lives. MLOps applies DevOps principles to machine learning: experiment tracking, model versioning, CI/CD for ML, monitoring, and incident response.

```python
import math

class MetricsTracker:
    def __init__(self, project, config):
        self.project = project
        self.config = config
        self.history = []
        print(f"=== Experiment: {project} ===")
        print(f"Config: {config}\n")

    def log(self, metrics):
        self.history.append(metrics)

    def summary(self):
        print(f"{'Step':>6s} | {'Loss':>8s} | {'LR':>10s} | {'Eval PPL':>10s}")
        print("-" * 42)
        for m in self.history:
            ppl = f"{m['eval/ppl']:>10.1f}" if 'eval/ppl' in m else f"{'---':>10s}"
            print(f"{m['step']:>6d} | {m['loss']:>8.4f} | {m['lr']:>10.6f} | {ppl}")

tracker = MetricsTracker("llm-training", {"model": "gpt-125m", "lr": 3e-4, "precision": "bf16"})

for step in range(10):
    loss = 4.0 * math.exp(-step * 0.15) + 0.05 * (step % 3)
    lr = 3e-4 * (1 - step / 20)
    metrics = {"step": step, "loss": loss, "lr": lr}
    if step % 3 == 0:
        metrics["eval/ppl"] = math.exp(loss * 1.1)
    tracker.log(metrics)

tracker.summary()
print(f"\nIn production: replace MetricsTracker with wandb.init() + wandb.log()")
```
