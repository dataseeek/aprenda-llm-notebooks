# Chapter 01 — Building a Bigram Model

Language models are the foundation of modern AI. At their core, they answer one deceptively simple question: given a sequence of text, what comes next? This chapter introduces you to language modeling through the simplest possible approach — the bigram model — so you can build intuition before tackling more complex architectures.

```python
import numpy as np

# Build frequency matrix from text
def build_bigram_matrix(text, vocab):
    size = len(vocab)
    char_to_idx = {c: i for i, c in enumerate(vocab)}
    matrix = np.zeros((size, size), dtype=np.float32)
    for ch1, ch2 in zip(text, text[1:]):
        matrix[char_to_idx[ch1]][char_to_idx[ch2]] += 1
    # Normalize rows to get probabilities
    row_sums = matrix.sum(axis=1, keepdims=True)
    matrix = matrix / np.maximum(row_sums, 1)
    return matrix, char_to_idx

# Generate text from the bigram model
def generate(matrix, char_to_idx, idx_to_char, seed='T', length=200):
    result = seed
    for _ in range(length):
        idx = char_to_idx[result[-1]]
        probs = matrix[idx]
        next_idx = np.random.choice(len(probs), p=probs)
        result += idx_to_char[next_idx]
    return result
```
