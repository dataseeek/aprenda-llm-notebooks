# Chapter 01 — Building a Bigram Model

Language models are the foundation of modern AI. At their core, they answer one deceptively simple question: given a sequence of text, what comes next? This chapter introduces you to language modeling through the simplest possible approach — the bigram model — so you can build intuition before tackling more complex architectures.

**Setup:** This notebook runs on CPU — no GPU or extra installs needed. Just NumPy + Python stdlib.

```python
import numpy as np
import urllib.request

def build_bigram_matrix(words, vocab):
    """Build a word-level bigram probability matrix."""
    size = len(vocab)
    word_to_idx = {w: i for i, w in enumerate(vocab)}
    matrix = np.zeros((size, size), dtype=np.float32)
    for w1, w2 in zip(words, words[1:]):
        matrix[word_to_idx[w1]][word_to_idx[w2]] += 1
    row_sums = matrix.sum(axis=1, keepdims=True)
    matrix = matrix / np.maximum(row_sums, 1)
    return matrix, word_to_idx

def generate(matrix, word_to_idx, idx_to_word, seed='A', length=50):
    """Generate text by sampling from bigram probabilities."""
    result = [seed]
    for _ in range(length):
        idx = word_to_idx.get(result[-1])
        if idx is None:
            break
        probs = matrix[idx]
        if probs.sum() == 0:
            break
        next_idx = np.random.choice(len(probs), p=probs)
        result.append(idx_to_word[next_idx])
    return ' '.join(result)

# --- Demo: Portuguese Wikipedia bigram generator ---
url = "https://raw.githubusercontent.com/dataseeek/aprenda-llm-notebooks/main/data/wikipedia_sample_pt.txt"
text = urllib.request.urlopen(url).read().decode("utf-8")

words = text.split()
vocab = sorted(set(words))
print(f"Corpus: {len(words):,} words | Vocab: {len(vocab):,} unique words\n")

matrix, word_to_idx = build_bigram_matrix(words, vocab)
idx_to_word = {i: w for w, i in word_to_idx.items()}

print("--- Texto Gerado (modelo bigram) ---")
print(generate(matrix, word_to_idx, idx_to_word, seed='A', length=50))
```
